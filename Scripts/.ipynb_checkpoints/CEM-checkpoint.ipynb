```python
# Listar arquivos no diretorio de arquivos #
lista = lista_arquivos(diretorio='./CSV', tipo=".csv")

# Separar entre conduzido e irradiado #
qtd_testes = 14
lcond = lista[0:qtd_testes]
lrad = lista[qtd_testes:2*qtd_testes]

# Abrir arquivos #
conduzida = np.array( abrir_arquivos(lcond, len(lcond), ',') )
radiada = np.array( abrir_arquivos(lrad, len(lrad), ',') )

# Leitura do nome dos arquivos para identificação no gráfico #
lcond_nome = []
lrad_nome = []
i = 0
while i < len(lcond):
    val = lcond[i].replace('./CSV/conduzida/','').replace('.csv','').replace('-','')
    val = re.sub(r"a\d+", "", val)
    lcond_nome.append(val)
    i += 1
    
i = 0
while i < len(lrad):
    val = lrad[i].replace('./CSV/radiada/','').replace('.csv','').replace('-','')
    val = re.sub(r"a\d+", "", val)
    lrad_nome.append(val)
    i += 1

# Valores da norma #
norma_cnome = 'CISPR22 classe B'#'EN55022 classe B'
norma_cfreq = [150e3, 500e3, 5e6, 5e6, 30e6]
norma_camp = [66, 56, 56, 60, 60]
norma_rnome = 'CISPR22 classe B'
norma_rfreq = [30e6, 230e6, 230e6, 1e9]
norma_ramp = [30, 30, 37, 37]
```

```python
params = { 'figure.figsize' : (20, 8),#(20/2.54, 8/2.54) ,
           'font.size' : 10,
           'lines.linewidth' : 1,
           'lines.color' : 'black',
           'axes.titlesize' : 10,
           'axes.labelsize' : 10,
           'axes.grid' : True,
           'axes.grid.which' : 'both',
           'grid.color' : 'silver',
           'grid.linestyle' : '-',
           'grid.linewidth' : 0.5,
           'xtick.labelsize' : 10,
           'ytick.labelsize' : 10,
           #'legend.loc' : 'lower right',
           'legend.fontsize': 10}

plt.rcParams.update(params)

# Dados do teste base #
base_cx, base_cy = get_xy_plot(conduzida[0])
base_rx, base_ry = get_xy_plot(radiada[0])

# Plotagem dos graficos com comparativo entre o teste base #
cor_atual = '#000099'
cor_anterior = '#9999CC'
cor_norma = '#555555'

i = 1
while i < qtd_testes:
    fig, ax = plt.subplots(1, 2)
    
    # --- Plotagem do grafico da conduzida --- #
    x, y = get_xy_plot(conduzida[i])
    ax[0].set_title('Conduzida')
    ax[0].set_ylabel('Amplitude [dB$\mu$V]')
    ax[0].set_xlabel('Frequência [Hz]')
    
    # Colocar a norma #
    ax[0].text(6e6, 60.5, norma_cnome, color=cor_norma) # 55022 Klasse B QPEAK
    ax[0].semilogx(norma_cfreq, norma_camp, color=cor_norma, ls=':' )
    
    ax[0].semilogx(base_cx, base_cy, label=lcond_nome[0], color=cor_anterior)
    ax[0].semilogx(x, y, label=lcond_nome[i], color=cor_atual)
    
    # Configuração dos eixos #
    ax[0].set_xlim(150e3, 30e6)
    ax[0].set_ylim(20, 80)
    ax[0].set_xticks([200e3, 500e3, 1e6, 2e6, 5e6, 10e6, 20e6])
    ax[0].get_xaxis().set_major_formatter(tck.EngFormatter())
    ax[0].legend(loc='upper right')
    
    # --- Plotagem do grafico da irradiada --- #
    x, y = get_xy_plot(radiada[i])
    ax[1].set_title('Irradiada')
    ax[1].set_ylabel('Amplitude [dB$\mu$V/m]')
    ax[1].set_xlabel('Frequência [Hz]')
    
    # Colocar a norma #
    ax[1].text(260e6, 37.2, norma_rnome, color=cor_norma)
    ax[1].semilogx(norma_rfreq, norma_ramp, color=cor_norma, ls=':' )
    
    ax[1].semilogx(base_rx, base_ry, label=lrad_nome[0], color=cor_anterior)
    ax[1].semilogx(x, y, label=lrad_nome[i], color=cor_atual)
    
    # Configuração dos eixos #
    ax[1].set_xlim(30e6, 1000e6)
    ax[1].set_ylim(10, 50)
    ax[1].set_xticks([30e6, 50e6, 100e6, 200e6, 500e6, 1000e6])
    ax[1].get_xaxis().set_major_formatter(tck.EngFormatter())
    ax[1].legend(loc='lower right')
    
    i += 1
```
